```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
```

```python
image = cv2.imread("/Users/ravshan/Documents/programming/DIP/images/image.png")
print(image.shape)

translation_matrix = np.float32([[1,0,-100], [0,1,100]])

translated_image = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0]))
print(translated_image.shape)
plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title("translated_image")
plt.show()
```

```python
(h,w) = image.shape[:2]
center = (w//2, h//2)

rotation_matrix = cv2.getRotationMatrix2D(center=center, angle=-90, scale=-1)

rotated_image = cv2.warpAffine(image, rotation_matrix, (h, w))

plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.show()

```

```python
f = 2
scaling_matrix = cv2.getRotationMatrix2D((0,0), 0, f)

scaled_image = cv2.warpAffine(image, scaling_matrix, (w*f, h*f))

plt.imshow(cv2.cvtColor(scaled_image, cv2.COLOR_BGR2RGB))
plt.show()
print(image.shape)
print(scaled_image.shape)
```
