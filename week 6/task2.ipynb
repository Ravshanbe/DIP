```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread("image.png")

kernel_3x3 = np.ones((3, 3), np.float32) / 9
blurred = cv2.filter2D(img, -1, kernel_3x3)

kernel_7x7 = np.ones((7,7), np.float32)/ (7*7)
blurred2 = cv2.filter2D(img, -1, kernel_7x7)

kernel_9x9 = np.ones((9,9), np.float32) / (9*9)
blurred3 = cv2.filter2D(img, -1, kernel_9x9)


plt.figure(figsize=[15,5])
plt.subplot(141);plt.imshow(img[...,::-1]);plt.title("Original")
plt.subplot(142);plt.imshow(blurred[...,::-1]);plt.title("blurred with 3x3 filter")
plt.subplot(143);plt.imshow(blurred2[...,::-1]);plt.title("blurred with 7x7 filter")
plt.subplot(144);plt.imshow(blurred3[...,::-1]);plt.title("blurred with 9x9 filter")
```

image details are getting lost as kernel size increases. Image gets blurry

```python
blur = cv2.blur(img, (7,7))

guassian = cv2.GaussianBlur(img, (3,3), 0)

median = cv2.medianBlur(img, 5)

bilateral = cv2.bilateralFilter(img, 9, 75, 75)

plt.figure(figsize=[15,10])
plt.subplot(231);plt.imshow(img[...,::-1]);plt.title("Original")
plt.subplot(232);plt.imshow(blur[...,::-1]);plt.title("After uniform filter")
plt.subplot(233);plt.imshow(guassian[...,::-1]);plt.title("After non-uniform gaussian filter")
plt.subplot(234);plt.imshow(median[...,::-1]);plt.title("After median filter")
plt.subplot(235);plt.imshow(bilateral[...,::-1]);plt.title("After bilateral filter")
```
