```python
grayscale_img = cv2.imread('image.png', cv2.IMREAD_GRAYSCALE)
cv2.imshow('grayscale image', grayscale_img)
cv2.imwrite("grayscaled_image_without_cvtcolor.png", grayscale_img)
# cv2.waitKey(0)
```

```python
image = cv2.imread("image.png")

grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imshow("grayscale image", grayscale_image)

# cv2.waitKey(0)

cv2.imwrite("grayscaled_image_with_cvtcolor.png", grayscale_image)
```

```python
import cv2

image = cv2.imread("image.png")

gr_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

_, binary_image = cv2.threshold(gr_image, 250, 255, cv2.THRESH_BINARY)

cv2.imwrite("grayscaled_image_with_threshhol_250.png", binary_image)
```
