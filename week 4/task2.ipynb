```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def contrast_stretching(img_path, s_max, s_min):

    img = cv2.imread(img_path)
    r_min = np.min(img)
    r_max = np.max(img)

    s = ((s_max - s_min)/(r_max - r_min))*(img-r_min) + s_min

    s = np.clip(s, s_min, s_max).astype('uint8')
    plt.figure(figsize=[15, 5])
    plt.subplot(121)
    plt.imshow(img[..., ::-1]) 
    plt.title("Original")
    plt.subplot(122)
    plt.imshow(s[..., ::-1])
    plt.title("Contrast Stretching")
    plt.show()

    return s
```

```python
res1 = contrast_stretching('1.png', 0, 50)
```
