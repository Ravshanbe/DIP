```python
import cv2

import cv2
image_color = cv2.imread('image.png')

b, g, r = cv2.split(image_color)

subtraction_value = 50

r_subtracted = cv2.subtract(r, subtraction_value)

modified_image = cv2.merge((b,g,r_subtracted))


cv2.imshow('Original Image', image_color)
cv2.imshow('Red Channel Subtracted', modified_image)
# cv2.waitKey(0)
```

```python
import cv2
import matplotlib.pyplot as plt
# Read a grayscale image
image = cv2.imread('image.png')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# Multiply an intensity value to all pixels
# multiplied_image = cv2.multiply(image_RGB, 2)
multiplied_image = cv2.multiply(image_RGB, 0.5)
#multiplied_image = cv2.multiply(image_RGB, 20)
# Display original and multiplied images
plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(multiplied_image)
plt.title('Multiplied Image 2')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
# Read a grayscale image
image = cv2.imread('image.png')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# Divide all pixels by an intensity value
multiplied_image = cv2.divide(image_RGB, 2)
# Display original and divided images
plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(multiplied_image)
plt.title('Divided Image 2')
plt.axis('off')
plt.show()
```
